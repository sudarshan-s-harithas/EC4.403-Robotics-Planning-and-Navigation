```python
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
class nonhn_RRT:
    
    def __init__(self,start_config,goal,v_init,w_init):
        (x0,y0,theta0) = start_config
        self.xg = goal[0]
        self.yg = goal[1]
        self.x = []
        self.y = []
        self.theta = []
        self.x.append(x0)
        self.y.append(y0)
        self.theta.append(theta0)
        self.vc = [] #platform centre v list
        self.wc = [] #platform centre w list
        self.vc.append(v_init)
        self.wc.append(w_init)
        
        self.map_height = [ 0 , 100] 
        self.map_width = [0 , 100 ]
        
        self.width_list = []
        self.height_list = []
        self.center_list = []
        self.numRectangles =0 
        self.RectCenter_list = []
        self.Radius_list = [] 
        self.map = np.zeros( ( self.map_width[1] , self.map_height[1]  ) )
        
        
        
        
        
    # implement non-holonomic RRT to find platform centre velocities
    
    
    def DetermineRectangeSizes( self, rectx , recty   ):
        
        numRects = len( rectx)/4
        
        widths = []
        heights = []
        prev_width =0
        prev_height = 0 
        center = []
        
        index = np.arange( 0 , len( rectx)  , 4 )
        j =0 
        
        for i in ( index ):
            
            w = rectx[i +2] - rectx[i]
            h = recty[i+2] - recty[i]
            
            c= [ (rectx[i +2] + rectx[i])/2 , (recty[i+2] + recty[i] )/2 ]
            
            widths.append(w)
            heights.append(h)
            center.append(c )
        
        return widths , heights , center , numRects
    
    
    def GetBoxCollisionSDF(self,  widths , heights , Qpt2 , numRects , RectCenter ,  collisionDistance ):
        
        self.width_list = widths
        self.height_list = heights
        self.numRectangles = numRects
        self.RectCenter_list = RectCenter
        
        
        
        for i in range( int(numRects) ):
            
            R = [  widths[i]/2  , heights[i]/2 ]
            Qpt = np.asarray(Qpt2)
            QptRectframe = Qpt - np.asarray(RectCenter[i] )
            Qpt = np.abs(QptRectframe)
#             print( Qpt )
#             print( RectCenter[i] )
#             print("---------")
            Temp = np.maximum( Qpt - R , np.zeros(2) )
            dist = np.linalg.norm(  Temp )
            collisionDistance.append(dist)
        
        return collisionDistance
    
    def GetSphericalCollision( self, CircleCenters  , collisionDistance , Qpt , radii ):
        
        numCircles = len(radii)
        self.Radius_list = radii
        self.center_list = CircleCenters
        
        for i in range(numCircles ):
            
            d = np.linalg.norm( CircleCenters[i] - Qpt  ) - radii[i]
            collisionDistance.append(d)
        
        return collisionDistance


    def StatePropagation( self, CurrState  ,control_u , dt ):
        
        X = CurrState[0] + control_u[0]*dt
        Y = CurrState[1] + control_u[1]*dt
        theta = CurrState[2] + control_u[2]*dt
        
        state = [X, Y,theta ]
        return state 
    
    def SampleStates(self ,CurrState , vel , angvel  , time_dt ):
        
        state = []
        th = CurrState[2] 
        CurrConfig2 = [ CurrState[0 ] , CurrState[1] , CurrState[2] ]
        cnt =0 
        append_val = True 
         
        for v in vel:
            for w in angvel:
                append_val = True 
                for t in time_dt:
                    collision = []
                    vx = v*math.cos(th)
                    vy = v*math.sin(th)
                    control_u = [ vx, vy, w]
                    [X, Y, th] = nn_RRT.StatePropagation(  CurrConfig2  ,control_u , t )
                    TempState = [X, Y ]
                    collision = self.GetBoxCollisionSDF(  self.width_list , self.height_list , TempState , self.numRectangles , 
                                                self.RectCenter_list ,  collision )
                    collision = self.GetSphericalCollision(self.center_list ,  collision , TempState , self.Radius_list )
                    minDist = np.amin(collision )
                    cnt +=1
#                     print(cnt)
                    if(minDist < 1 ):
                        append_val = False 
                
                if( append_val):
                    state.append( [X, Y, th] )
                    
        
        
        return state 
        
    def SampleRandomConfig(self, CurrConfig ):  ## Sample Local Goal 
        
        isPtValid = False  
        
        while not isPtValid:
            collision2 = []
            xval = np.random.randint( low = self.map_width[0] , high = self.map_width[1]   )
            yval = np.random.randint( low = self.map_width[0] , high = self.map_width[1]   )
            TempState = [xval ,  yval ]
            collision2 = self.GetBoxCollisionSDF(  self.width_list , self.height_list , TempState , self.numRectangles , 
                                                self.RectCenter_list ,  collision2 )
            collision2 = self.GetSphericalCollision(self.center_list ,  collision2 , TempState , self.Radius_list )
            minDist = np.amin(collision2 )
            if( minDist > 10 ):
                isPtValid = True  
 
                
            
        return TempState
    
    
    def EstimateBestControl(  self, Prostates , LocalGoal ):
        
        DistToLocalGoal = []
        for pro_state in Prostates:
            Qpt =[ pro_state[0] , pro_state[1] ]
            dist = np.linalg.norm( np.asarray(LocalGoal) - np.asarray( Qpt )  )
            DistToLocalGoal.append(dist )
#             print(dist)
        
        index = np.argmin(DistToLocalGoal )
        
        return index 
             
            
    
    def showpath(self):
        """
        Create a new directory (say nonholonomic_path) and save snapshots of the robot moving along the
        trajectory found by RRT at every step with the step index.
        e.g: snap1.png, snap2.png ...
        """
        
        
    def showtree(self):
        """
        Create a new directory (say nonholonomic_tree) and save snapshots of the evolution of the RRT tree
        at every step with the step index.
        e.g: snap1.png, snap2.png ...
        """
        
    def test_env(self):
        """
        Function to generate test environment
        """
        # the vertices for each obstacle are given in the clockwise order starting from lower left
        obst_vertex_x = [40,40,60,60,70,70,80,80,40,40,60,60] 
        obst_vertex_y = [62,100,100,62,40,60,60,40, 0,38,38, 0]
        plt.plot(self.xg,self.yg,'g*',markersize=20)
        plt.plot(self.x[0],self.y[0],'b*',markersize=20)
        for i in range(1,4): 
            x = obst_vertex_x[4*(i-1)]
            y = obst_vertex_y[4*(i-1)]
            width = obst_vertex_x[4*(i-1)+2] - obst_vertex_x[4*(i-1)]
            height = obst_vertex_y[4*(i-1)+2] - obst_vertex_y[4*(i-1)]
            rect = plt.Rectangle((x,y),width,height,linewidth=1,color='r')
            plt.gca().add_patch(rect)
           
        #centre positions of circular obstacles
        circle_x = [20,90,77]
        circle_y = [20,70,82]
        radii = [7,2,3]

        for i in range(3): #plotting  circles
            circle = plt.Circle((circle_x[i],circle_y[i]),radii[i], color='r')
            plt.gca().add_patch(circle)

        plt.gca().set_aspect('equal', adjustable='box')
        plt.show()
        
        return obst_vertex_x , obst_vertex_y , circle_x , circle_y , radii
    
    
    def PlotPaths( self,   ProStates):
        obst_vertex_x = [40,40,60,60,70,70,80,80,40,40,60,60] 
        obst_vertex_y = [62,100,100,62,40,60,60,40, 0,38,38, 0]
        plt.plot(self.xg,self.yg,'g*',markersize=20)
#         plt.plot(self.x[0],self.y[0],'b*',markersize=20)
        for i in range(1,4): 
            x = obst_vertex_x[4*(i-1)]
            y = obst_vertex_y[4*(i-1)]
            width = obst_vertex_x[4*(i-1)+2] - obst_vertex_x[4*(i-1)]
            height = obst_vertex_y[4*(i-1)+2] - obst_vertex_y[4*(i-1)]
            rect = plt.Rectangle((x,y),width,height,linewidth=1,color='r')
            plt.gca().add_patch(rect)
           
        #centre positions of circular obstacles
        circle_x = [20,90,77]
        circle_y = [20,70,82]
        radii = [7,2,3]
        X_val = []
        Y_val = [] 

        for i in range(3): #plotting  circles
            circle = plt.Circle((circle_x[i],circle_y[i]),radii[i], color='r')
            plt.gca().add_patch(circle)
            
        for State in ProStates:
            X_val.append( State[0] )
            Y_val.append( State[1] )
        
        plt.plot( X_val ,Y_val  )
            

        plt.gca().set_aspect('equal', adjustable='box')
        plt.show()
        
```

```python
class Utilities:
    
    def SampleControls(  Vrange, Wrange , NumCntrSamples ):
        
        V = []
        W = []
        
        delv = (Vrange[1] - Vrange[0])/NumCntrSamples
        delw = (Wrange[1] - Wrange[0])/NumCntrSamples
        
        
        for i in range(NumCntrSamples +1 ):
            vtemp = Vrange[0]+ i*delv
            wtemp = Wrange[0] + i*delw
            
            V.append( vtemp)
            W.append(wtemp)
        
        
        return V, W 
    
    def GetParent( AllStates , QState):
        
        numStates = len( AllStates)
        dist =[]
        cnt = 0 
        for st in AllStates:

            
            d = np.linalg.norm( np.asarray( QState) - np.asarray( st ) )
            dist.append(d)
        
        index = np.argmin(dist)
        
        return AllStates[index]
            
            
            
            
```

```python
goal = (100,100)
start_config =(0,1,0)
vinit = 5
winit = 0.1
collisionDistance = [] 
nn_RRT = nonhn_RRT(start_config,goal,vinit,winit)
rectx , recty , circle_centerx , circle_centery , radii = nn_RRT.test_env()
CircleCenters = np.zeros( ( len(circle_centerx) , 2 ) )
CircleCenters[: , 0 ] = circle_centerx
CircleCenters[: , 1] = circle_centery

Vrange = [ -3, 3 ]
Wrange = [-3 , 3 ]
NumCntrlSamples = 5
T = 4 
steps = 4
dt = T/steps 
time_dt = np.arange( 0 , T, dt) 
```

```python
CurrConfig = [0 ,0   ]

CurrConfig2 = [10.5,10,10]


widths , heights , center , numRects  = nn_RRT.DetermineRectangeSizes( rectx , recty )
collisionDistance = nn_RRT.GetBoxCollisionSDF( widths , heights , CurrConfig , numRects , center ,  collisionDistance )
collisionDistance = nn_RRT.GetSphericalCollision( CircleCenters  , collisionDistance , CurrConfig , radii )
vel , angvel = Utilities.SampleControls(Vrange ,Wrange ,NumCntrlSamples  )
state = nn_RRT.SampleStates( CurrConfig2 , vel , angvel  , time_dt )
LocalGoal = nn_RRT.SampleRandomConfig( CurrConfig2 )
BestControl = nn_RRT.EstimateBestControl( state , LocalGoal)

print(widths , heights , center , numRects)   
```
