```python
def DetermineBernsteinCoeffsX( x0 , xf , xc  , dx0, dx_f , dxc , t0, tc , tf  ):
    
    t0,t,tf = t0, tc,tf
    Wx0 = x0
    Wx5 = xf
    
    
    
    Bc0,Bc1,Bc2,Bc3,Bc4,Bc5 = get_bernstein_poly( t0 , tc , tf )
    Bc0dot, Bc1dot, Bc2dot, Bc3dot, Bc4dot, Bc5dot = get_bernstein_differentials( t0, tc , tf)
    B00dot, B01dot, B02dot, B03dot, B04dot, B05dot = get_bernstein_differentials( t0, t0 , tf)
    Bf0dot, Bf1dot, Bf2dot, Bf3dot, Bf4dot, Bf5dot = get_bernstein_differentials( t0, tf , tf)
    
    Ax = np.array( [ [ xc - Wx0*Bc0 -Wx5*Bc5  ]  , 
                     [dx0 - Wx0*B00dot - Wx5*B05dot  ] , 
                     [ dx_f - Wx0*Bf0dot - Wx5*Bf5dot ] ] ) 
#                      [dxc - Wx0*Bc0dot - Wx5*Bc5dot ] ])
    
    Bx = np.array( [ [ Bc1,Bc2,Bc3,Bc4  ] , 
                     [ B01dot, B02dot, B03dot, B04dot ] , 
                     [Bf1dot, Bf2dot, Bf3dot, Bf4dot   ] ])  
#                      [ Bc1dot, Bc2dot, Bc3dot, Bc4dot ] ] )
    W = np.linalg.inv( Bx.T@Bx + np.eye(4)*0.002 )@Bx.T@Ax   #np.linalg.inv( Bx)@Ax
    
    
    Wx = np.array( [Wx0 , W[0][0] ,W[1][0], W[2][0] , W[3][0] , Wx5  ])
    
    return Wx    
```

```python
def DetermineBernsteinCoeffsK( k0, kf, kc ,  kdott0 , kdottf , yt0, ytc , ytf , t0 , tf , Wx , tc ):
    
    Wk0 = k0
    Wkf = kf
    
#     print( t0 , tf, tc )
    
#     t0 , tf, tc  = 0 , 50 , 10
    
#     print( t0 , tf, tc )
    
    
    
    
    coef0t0, coef1t0, coef2t0, coef3t0, coef4t0, coef5t0 = get_product_functions( Wx[0] ,Wx[1]
                                                                                 ,Wx[2],Wx[3],Wx[4],Wx[5],
                                                                                 t0,t0,tf )
    coef0tf, coef1tf, coef2tf, coef3tf, coef4tf, coef5tf = get_product_functions( Wx[0] ,Wx[1]
                                                                                 ,Wx[2],Wx[3],Wx[4],Wx[5],
                                                                                 t0,tf,tf )
    coef0tc, coef1tc, coef2tc, coef3tc, coef4tc, coef5tc = get_product_functions( Wx[0] ,Wx[1]
                                                                                 ,Wx[2],Wx[3],Wx[4],Wx[5],
                                                                                 t0,tc,tf )  
    
    
    B00dot, B01dot, B02dot, B03dot, B04dot, B05dot = get_bernstein_differentials( t0, t0 , tf)
    Bf0dot, Bf1dot, Bf2dot, Bf3dot, Bf4dot, Bf5dot = get_bernstein_differentials( t0, tf , tf)
    
    
    Bc0,Bc1,Bc2,Bc3,Bc4,Bc5 = get_bernstein_poly( t0 , tc , tf )
    B00,B01,B02,B03,B04,B05 = get_bernstein_poly( t0 , t0 , tf )
    Bf0,Bf1,Bf2,Bf3,Bf4,Bf5 = get_bernstein_poly( t0 , tf , tf )
    
#     print( k0, kf , yt0,ytf ,kc, ytc )
    
#     k0, kf , yt0,ytf ,kc, ytc = 0 ,0,0,40,1,40
    
#     print( k0, kf , yt0,ytf ,kc, ytc )
    
#     print( Bc0,Bc1,Bc2,Bc3,Bc4,Bc5 )
    
    BC = np.array( [ Bc0,Bc1,Bc2,Bc3,Bc4,Bc5 ])

    
    Ak = np.array( [ [k0 -Wk0*B00 -  Wkf*B05 ] , 
                   [  kf -Wk0*Bf0 - Wkf*Bf5   ]  , 
                   [ yt0 - Wk0*coef0t0 -  Wkf*coef5t0  ] , 
                   [ ytf - Wk0*coef0tf -  Wkf*coef5tf  ] , 
                   [  kc - Wk0*Bc0 - Wkf*Bc5 ] , 
                   [ ytc - Wk0*coef0tc - Wkf*coef5tc  ]])
    
    Bk = np.array( [ [ B01,B02,B03,B04 ] , 
                   [ Bf1,Bf2,Bf3,Bf4 ] , 
                   [ coef1t0, coef2t0, coef3t0, coef4t0 ] , 
                    [coef1tf, coef2tf, coef3tf, coef4tf ], 
                    [Bc1,Bc2,Bc3,Bc4],
                    [ coef1tc, coef2tc, coef3tc, coef4tc ] 
                   ])
    
    
#     Ak = np.array([ [ yt0 - Wk0*coef0t0 -  Wkf*coef5t0   ] , 
#                     [ ytf - Wk0*coef0tf -  Wkf*coef5tf ]  , 
#                     [ ytc - Wk0*coef0tc - Wkf*coef5tc  ] , 
#                     [ kc - Wk0*Bc0 - Wkf*Bc5  ]])
    
#     Bk = np.array( [[ coef1t0, coef2t0, coef3t0, coef4t0 ] , 
#                    [ coef1tf, coef2tf, coef3tf, coef4tf  ] , 
#                    [coef1tc, coef2tc, coef3tc, coef4tc ] , 
#                    [ Bc1,Bc2,Bc3,Bc4  ]])
    
    W = np.linalg.inv(Bk.T@Bk  + np.eye(4)*0.000002  )@(Bk.T)@Ak  # + np.eye(4)*0.000002
    
    
    
    Wk = np.asarray([ Wk0 , W[0][0] , W[1][0] , W[2][0] , W[3][0] , Wkf ]) 
    

    
    
    return Wk
```

```python
def GetTheta( Wk , t0, tf , numPts , ObsPath , CollisionTime ):
    
#     Wk0, Wk1, Wk2, Wk3, Wk4, Wk5 = Wk[0] , Wk[1] , Wk[2] , Wk[3] , Wk[4] , Wk[5]
    
    delval = ( tf -t0)/(numPts)
#     Wk = np.reshape( Wk , (1, 6) )
    
    T = []
    
    for i in range( numPts+1  ):
        
        T.append( t0 + i*delval )
    
    T = np.asarray(T)
    
    Theta = []
    Theta_Deg = []
    
    for i in range( numPts +1):
        
        B0,B1,B2,B3,B4,B5 = get_bernstein_poly(t0,T[i],tf)
        B =   np.array( [ B0,B1,B2,B3,B4,B5 ]) 
        theta = np.arctan( Wk@B)
        theta_deg = theta*180/3.14 
        Theta_Deg.append( theta_deg )
        Theta.append(theta)

    
    return Theta , Theta_Deg
```

```python
def GetDirection(theta  ):
    
    length = len(theta)
    
    x_direction = []
    y_direction = []
    
    for i in range( length ):
        
        x_dir = np.cos(theta[i])
        y_dir = np.sin(theta[i])
        x_direction.append( x_dir )
        y_direction.append(  y_dir )
    
    
    return x_direction , y_direction
```

```python
def GetObstaclePath( startPt , collisionPt , CollisionTime , t0, tf , numPts ): 
    
    ## Input the start point of the dynamic obstacle and the collision place and time 
    
    length_to_cover  = np.linalg.norm(startPt - collisionPt )
    velocity_magnitude = length_to_cover/CollisionTime
    
    velocity_direction = (collisionPt - startPt)/length_to_cover
    
    delval = ( tf -t0)/(numPts)
    
    T = []
    
    for i in range( numPts+1  ):
        T.append(t0 + i*delval )
    
    
    obstacle_trajectory = [] 
    
    
    for i in range( numPts+1  ):
        
        Pt = startPt + (velocity_direction*velocity_magnitude)*T[i]
        obstacle_trajectory.append(Pt)
    
    
    return obstacle_trajectory
```

```python
def PerformTimeScaling( Wx , Wk , y0,  t0, tf , startPt , collisionPt , CollisionTime , Obsradius  ):
    
    t = t0
    delt = 2.5 
    cnt = 0 
        
    Wx0,Wx1,Wx2,Wx3,Wx4,Wx5 = Wx[0] , Wx[1] , Wx[2] , Wx[3] , Wx[4] , Wx[5]
    
    length_to_cover  = np.linalg.norm(startPt - collisionPt )
    velocity_magnitude = length_to_cover/CollisionTime
    
    velocity_direction = (collisionPt - startPt)/length_to_cover

    Theta_Deg = []
    Theta = [] 
    
    Xpts = []
    Ypts = []
    
    ObsPath  = []
    Time = []
    go_here_from_on= False 
    enter_once = True
    
    end_time = tf +delt
#     scale = 2
    
    
    Obs2StartPt = np.array([40, -100])
    Collision2Pt = np.array([24,-163.166])
    
    Collision2Time  = 37.5
    
    len_to_cover2 = np.linalg.norm(Obs2StartPt - Collision2Pt )
    velocity_magnitude2 = len_to_cover2/Collision2Time
    
    velocity_direction2 = (Collision2Pt - Obs2StartPt)/len_to_cover2
    Obs2Path = []
    
    
    while ( t < end_time ):
        
        
            
        B0,B1,B2,B3,B4,B5 = get_bernstein_poly(t0,t,tf)
        B = np.array( [ B0,B1,B2,B3,B4,B5 ])
        xpt = Wx@B
        
        coef0t, coef1t, coef2t, coef3t, coef4t, coef5t = get_product_functions( Wx0,Wx1,Wx2,Wx3,Wx4,Wx5,t0,t,tf )
        F = np.array( [coef0t, coef1t, coef2t, coef3t, coef4t, coef5t ])
        ypt = y0 + Wk@F
        
        X = np.array( [ xpt ,ypt ])
        
        ObsPt = startPt + (velocity_direction*velocity_magnitude)*t
        ObsPath.append(ObsPt  )
        
        Obs2Pt = Obs2StartPt + velocity_direction2*velocity_magnitude2*t
        Obs2Path.append( Obs2Pt)
        
        dist_to_obs1 = np.linalg.norm( ObsPt - X ) - Obsradius 
        dist_to_obs2 = np.linalg.norm( Obs2Pt - X ) - Obsradius 
        
        
        dist_to_obs = min(dist_to_obs1 , dist_to_obs2 )
        
        if( dist_to_obs < 1   ):
            print( "in time " + str( t) )         
            scale = 2 #GetScale(  [ Xpts[-1] , Ypts[-1] ]   , ObsPath[-2] , collisionPt , delt , Obsradius )
            
#             print(scale)
#             scale= 1
#             go_here_from_on = True 
#             enter_once = True 
            
            if( enter_once ):
                enter_once = True
                t -= delt
            
            t += scale*delt 
            
            if( t > end_time ):
                t = end_time
            B0,B1,B2,B3,B4,B5 = get_bernstein_poly(t0,t,tf)
            B = np.array( [ B0,B1,B2,B3,B4,B5 ])
            coef0t, coef1t, coef2t, coef3t, coef4t, coef5t = get_product_functions( Wx0,Wx1,Wx2,Wx3,Wx4,Wx5,t0,t,tf )
            F = np.array( [coef0t, coef1t, coef2t, coef3t, coef4t, coef5t ])
#             t += scale*delt
            xpt = Wx@B
            ypt = y0 + Wk@F
            
        theta = np.arctan( Wk@B)
        theta_deg = theta*180/3.14 
        Theta_Deg.append( theta_deg )
        Theta.append(theta)
        Xpts.append(xpt  )
        Ypts.append(ypt )
        
#         print(t)
        
                
        if(go_here_from_on == False ):
            t += delt
            
        Time.append(2.5*cnt)
        cnt += 1 
    
    return Xpts , Ypts ,Theta , Theta_Deg  ,ObsPath, Time , Obs2Path
```

```python
class agent:
    def __init__(self,rad,tf,tw,start,waypt,goal,init_or,waypt_or,final_or,dx_0,dx_f,dx_w):
        
        self.init_or = init_or
        self.final_or = final_or
        self.waypt_or = waypt_or
        xdoto=0  
        xdotf=0
        self.k0=np.tan(self.init_or)   # tangent of the initial heading angle of the robot
        self.kdot0=0 # differntial of the tangent of the initial heading angle of the robot
        self.kw = np.tan(self.waypt_or)
        self.kdotw = 0
        self.kf=np.tan(self.final_or)     #  tangent of the final heading angle of the robot
        self.kdotf=0  # differntial of the tangent of the final heading angle of the robot
        self.rad = rad
        self.t0 = 0
        self.tf = tf
        self.tw = tw
        self.pos = start
        self.x = []
        self.y = []
        self.xdot = []
        self.ydot = []
        self.goal = goal
        self.origin = start
        self.waypt = waypt
        self.xf = self.goal[0]
        self.yf = self.goal[1]
        self.dx_0 = dx_0
        self.dx_f = dx_f
        self.dx_w = dx_w
        
        self.x0 = self.pos[0]
        self.y0 = self.pos[1]
        self.xc = self.waypt[0]
        self.yc = self.waypt[1]
        
        self.dxc = (self.dx_0 +self.dx_f )/2 + 3
        self.dyc = (self.dx_f +self.dx_f )/2
        
        self.tc = self.tw
        self.Obsradius = 2

                    
    def get_overall_traj(self):
         
        Wx = DetermineBernsteinCoeffsX(self.x0 , self.xf , self.xc  , self.dx_0, self.dx_f ,  self.dxc , self.t0 ,self.tc  ,  self.tf )
#         print(Wx)
        
        Wk = DetermineBernsteinCoeffsK( self.k0 , self.kf , self.kw  ,  self.dx_w , self.dx_w , 
                                       self.pos[1] , self.waypt[1]  , self.goal[1] , self.t0 , self.tf , Wx , self.tc )
          
        
#         theta  , Theta_Deg = GetTheta( Wk , self.t0, self.tf , 20    )
        
        
        collisionPt = np.array([ 6.004627180232551 , 36.73269189312435 ])
        CollisionTime = 12.5
        ObsStartPt = np.array([-10 , 10])
        
#         ObsPath = GetObstaclePath( ObsStartPt , collisionPt , CollisionTime , self.t0, self.tf , 20 )
        
#         Xpts = GetXPts( Wx , self.t0, self.tf , 20 , ObsPath , CollisionTime )
#         Ypts , Time = GetYpts( self.y0  , Wk,self.t0, self.tf , 20 , Wx , ObsPath , CollisionTime)
#         theta  , Theta_Deg = GetTheta( Wk , self.t0, self.tf , 20   , ObsPath  , CollisionTime )
#         x_direction , y_direction = GetDirection(  theta  )
        
        Xpts  , Ypts , theta, Theta_Deg  ,ObsPath , Time ,Obs2Path = PerformTimeScaling( Wx , Wk ,self.y0,  self.t0, self.tf , ObsStartPt ,  collisionPt
                                                 , CollisionTime , self.Obsradius )
        x_direction , y_direction = GetDirection(  theta  )
        
        print( Xpts , Ypts )
        
        
        
        
        plt.plot(Time , Xpts , label ="X-axis" )
        plt.plot(Time , Ypts , label ="Y-axis")
        plt.plot( Time ,  Theta_Deg  , label ="Orientation" )
        plt.xlabel("Time")
        plt.legend()
        plt.show()            
        vis_trajectory( Xpts, Ypts , x_direction , y_direction  ,ObsPath ,  Time , self.Obsradius , Obs2Path )
        

              
    def visualize_traj(self,time):
        """
        Function to visualize robot trajectory at a given timestamp. 
        Saves snapshots to /data folder as snap0.png, snap1.png...
        String the snapshots using the bash script mkmovie.sh to generate simulation video.
        """
        figure = plt.figure()
        ax = figure.add_subplot(1,1,1)
        robot = matplotlib.patches.Circle(
            (self.pos[0],self.pos[1]),
            radius = self.rad,
            edgecolor='black',
            linewidth=1.0,
            ls='solid',
            alpha=1,
            zorder=2)
        ax.add_patch(robot)
        name = 'data/snap%s.png'%str(time)
        ax.plot([self.goal[0]], [self.goal[1]], '*', color="red", markersize =15,linewidth=3.0)
        ax.plot([self.waypt[0]], [self.waypt[1]], '*', color="green", markersize =15,linewidth=3.0)
        ax.plot(self.x,self.y,"b--")
        ax.set_aspect('equal')
        ax.set_xlim(-10.0, 60.0)
        ax.set_ylim(-10.0, 60.0)
        ax.set_xlabel(r'$x (m)$')
        ax.set_ylabel(r'$y (m)$')
        ax.grid(True)
        plt.savefig(name, dpi = 200)
        plt.cla()
        plt.close(figure)
        return figure
```
